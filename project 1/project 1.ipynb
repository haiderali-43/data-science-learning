```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

Drop unnamed column

```python
df = df.drop("Unnamed: 0", axis=1)
print(df.head())
```

```python
df.head()
```

Gender distribution


```python
ax = sns.countplot(data = df , x = 'Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean", "WritingScore":"mean", "ReadingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb, annot=True)
plt.show()
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean", "WritingScore":"mean", "ReadingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1, annot=True)
plt.show()
```

```python
sns.boxplot(data=df, x='MathScore')
plt.show()
```

```python
print(df["EthnicGroup"].unique())
```

```python
groupA = df.loc[(df['EthnicGroup']=='group A')].count()
groupB = df.loc[(df['EthnicGroup']=='group B')].count()
groupC = df.loc[(df['EthnicGroup']=='group C')].count()
groupD = df.loc[(df['EthnicGroup']=='group D')].count()
groupE = df.loc[(df['EthnicGroup']=='group E')].count()

l = ['groupA', 'groupB', 'groupC', 'groupD', 'groupE']
mylist = [groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
plt.pie(mylist, labels=l, autopct='%1.2f%%')
plt.show()
```
